```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from mlxtend.plotting import plot_decision_regions
```

# Bagging and Pasting

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.show()
```

```python
plt.hist(y_train)
plt.hist(y_test)
plt.show()
```

* Scikit-Learn oferuje prosty interfejs API do **bagging** i **pasting** . 

* Poniszy kod uczy zesp zoony z 500 drzewa decyzyjnego, z ktre zostao nauczonych na 100 elementach ze zbioru uczcego losowo pobranych z zestawu treningowego ze zwracaniem 

* Jest to przykad **bagging**, ale jeli chcesz uy **pasting** zamiast tego, po prostu ustaw **bootstrap=False**. 

* Parametr **n_jobs** mwi Scikit-Learn ile rdzeni procesora mona uy do treningu i prognoz (-1 mwi Scikit-Learn, aby uywa wszystkich dostpnych rdzeni):

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(random_state=42), n_estimators=500,
    max_samples=100, bootstrap=True, n_jobs=-1, random_state=42)
bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
```

The BaggingClassifier automatycznie wykonuje mikkie gosowanie zamiast twardego, jeli klasyfikator bazowy moe oszacowa prawdopodobiestwa klasy (tj. jeli ma metod **predict_proba()**), co ma miejsce w przypadku klasyfikatorw Decision Trees.

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)
y_pred_tree = tree_clf.predict(X_test)
print(accuracy_score(y_test, y_pred_tree))
```

# Zad. 

Narysuj wynik dla

* BaggingClassifier
* DecisionTreeClassifier

```python
for clf in (bag_clf,tree_clf):
    plt.figure(figsize=(8, 5))
    plot_decision_regions(X_test, y_test, clf)
    plt.show()
```

# Out-of-Bag Evaluation

* W przypadku metody **bagging** niektre przykady mog by prbkowane kilka razy dla dowolnego predyktora  podczas gdy inne mog nie by prbkowane w ogle. 

* Domylnie BaggingClassifier pobiera prbki ze zwracaniem (**bootstrap = True**), gdzie $m$ jest rozmiarem zestawu treningowego. 

* Oznacza to, e tylko okoo 63% instancji szkoleniowych jest prbkowanych rednio dla kadego z predyktorw. 

* Pozostae 37% instancji szkoleniowych, ktre nie s prbkowane, nazywane s instancjami **out-of-bag (oob)**. 

* Zauwa, e nie s one takie same dla wszystkich predyktorw. 

* Poniewa predykator nigdy nie widzi instancji **out-of-bag (oob)** podczas treningu, klasyfikator moe by oceniany na tych instancjach. 

* Moesz oceni komitet klasyfikatorw uredniajc oceny kadego z predyktorw na **out-of-bag (oob)**. 

```python
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), 
    n_estimators=500,
    bootstrap=True, 
    n_jobs=-1, 
    oob_score=True)

bag_clf.fit(X_train, y_train)
bag_clf.oob_score_
```

```python
from sklearn.metrics import accuracy_score
y_pred = bag_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```
