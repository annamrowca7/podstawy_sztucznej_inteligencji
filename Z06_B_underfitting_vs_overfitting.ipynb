# Underfitting vs. Overfitting

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn import  metrics
```

Rozważmy następujący zbiór punktów:

```python
np.random.seed(0)
n_samples = 30
true_fun = lambda X: np.cos(1.5 * np.pi * X)
X = np.sort(np.random.rand(n_samples))
y = true_fun(X) + np.random.randn(n_samples) * 0.1
```

```python
# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X, y,  color='black')
x_tr = np.linspace(0, 1, 200)
plt.show()
```

Podziel dane na dwie grupy.

```python
s=np.random.random_sample(n_samples)
s[s>0.5]=1
s[s<=0.5]=0
s
X1=X[s==1]
y1=y[s==1]
X2=X[s==0]
y2=y[s==0]

# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X1, y1,  color='blue')
plt.scatter(X2, y2,  color='red')
x_tr = np.linspace(0, 1, 200)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import  metrics
import sklearn.linear_model as lm
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
X1=np.vstack(X1)
model1 = linear_model.LinearRegression()
model1.fit(X1, y1)

plt.figure(figsize=(6,6))
axes = plt.gca()
axes.set_ylim([-1.5,1.5])
axes.set_xlim([0,1])
plt.scatter(X1, y1,  color='black')
plt.plot(X1, model1.predict(X1), color='blue',linewidth=3)
plt.show()
```

```python
X2=np.vstack(X2)
model1 = linear_model.LinearRegression()
model1.fit(X2, y2)

plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_ylim([-1.5,1.5])
axes.set_xlim([0,1])
plt.scatter(X2, y2,  color='black')
plt.plot(X2, model1.predict(X2), color='blue',linewidth=3)
plt.show()
```

```python
model2x1 = make_pipeline(PolynomialFeatures(20), linear_model.LinearRegression())
model2x1.fit(X1, y1)
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_ylim([-1.5,1.5])
axes.set_xlim([0,1])
plt.scatter(X1, y1,  color='black')
x_plot = np.vstack(np.linspace(0, 10, 100))
plt.plot(x_plot, model2x1.predict(x_plot), color='blue',linewidth=3)
plt.show()
```

```python
model2x2 = make_pipeline(PolynomialFeatures(20), linear_model.LinearRegression())
model2x2.fit(X2, y2)
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_ylim([-1.5,1.5])
axes.set_xlim([0,1])
plt.scatter(X2, y2,  color='black')
x_plot = np.vstack(np.linspace(0, 10, 100))
plt.plot(x_plot, model2x2.predict(x_plot), color='blue',linewidth=3)
plt.show()
```
